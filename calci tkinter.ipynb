```python
# import module and instance object
from tkinter import *
#create a GUI window
root = Tk()
# set the background colour of GUI window
root.configure(background = 'black')
#Window title
root.title("Simple Calculator")
# Window icon
root.iconbitmap('logoico.ico')

#create the text entry box for 
# showing the expression . 
e = Entry(root,bd = 10,font="Times 20",justify=RIGHT) 
# Grid Method is used for placing the widget at respective place
e.grid(row=0,column=0,columnspan=4,pady=3,sticky=W+E)
def button(num):
    current = e.get()
    e.delete(0, END)
    con = str(current) + str(num)
    e.insert(0,con)
    
def add():
    global frist_num
    first_num = e.get()
    global f_num
    global operator
    operator = "+"
    f_num = int(first_num)    
    e.insert(len(first_num), "+")
def sub():
    first_num = e.get()
    global f_num
    global operator
    operator = "-"
    f_num = int(first_num)    
    e.insert(len(first_num), "-")
def mul():
    first_num = e.get()
    global f_num
    global operator
    operator = "*"
    f_num = int(first_num)    
    e.insert(len(first_num), "*")
def div():
    first_num = e.get()
    global f_num
    global operator
    operator = "/"
    f_num = int(first_num)    
    e.insert(len(first_num), "/")
def clear():
    e.delete(0, END)
    
def equal():
    global first_num
    global operator
    if operator == "+":
        sec_num = (e.get()).split("+")[1]
        e.delete(0, END)
        e.insert(0, f_num + int(sec_num))
    if operator == "-":
        sec_num = (e.get()).split("-")[1]
        e.delete(0, END)
        e.insert(0, f_num - int(sec_num))
    if operator == "*":
        sec_num = (e.get()).split("*")[1]
        e.delete(0, END)
        e.insert(0, f_num * int(sec_num))
    if operator == "/":
        sec_num = (e.get()).split("/")[1]
        e.delete(0, END)
        e.insert(0, f_num / int(sec_num))
# Creating Buttons (Numbers)
# if we need pass any arguments throught a funtion we need to use lambda 

btn_1 = Button(root, text = "1",padx=30,pady=15 ,
               command = lambda : button(1),relief=RIDGE,bd=2).grid(row=3,column=0,padx=2,pady=2)
btn_2 = Button(root, text = "2",padx=30,pady=15 , 
               command = lambda : button(2),relief=RIDGE,bd=2).grid(row=3,column=1,padx=2,pady=2)
btn_3 = Button(root, text = "3",padx=30,pady=15 , 
               command = lambda : button(3),relief=RIDGE,bd=2).grid(row=3,column=2,padx=2,pady=2)

btn_4 = Button(root, text = "4",padx=30,pady=15 , 
               command = lambda : button(4),relief=RIDGE,bd=2).grid(row=2,column=0,padx=2,pady=2)
btn_5 = Button(root, text = "5",padx=30,pady=15 , 
               command = lambda : button(5),relief=RIDGE,bd=2).grid(row=2,column=1,padx=2,pady=2)
btn_6 = Button(root, text = "6",padx=30,pady=15 , 
               command = lambda : button(6),relief=RIDGE,bd=2).grid(row=2,column=2,padx=2,pady=2)

btn_7 = Button(root, text = "7",padx=30,pady=15,  
               command = lambda : button(7),relief=RIDGE,bd=2).grid(row=1,column=0,padx=2,pady=2)
btn_8 = Button(root, text = "8",padx=30,pady=15 , 
               command = lambda : button(8),relief=RIDGE,bd=2).grid(row=1,column=1,padx=2,pady=2)
btn_9 = Button(root, text = "9",padx=30,pady=15 , 
               command = lambda : button(9),relief=RIDGE,bd=2).grid(row=1,column=2,padx=2,pady=2)

btn_0 = Button(root, text = "0",padx=30,pady=15 , 
               command = lambda : button(0),relief=RIDGE,bd=2).grid(row=4,column=1,padx=2,pady=2)
# Creating Buttons (Sign)
# Here No need to use lambda (We are just click the button and nothing wanna pass any value)
btn_add = Button(root, text = "+",padx=30,pady=15 ,
                 command = add,relief=RIDGE,bd=2).grid(row=1,column=3,padx=2,pady=2)
btn_sub = Button(root, text = "-",padx=30,pady=15 , 
                 command = sub,relief=RIDGE,bd=2).grid(row=2,column=3,padx=2,pady=2)
btn_mul = Button(root, text = "*",padx=30,pady=15 ,
                 command = mul,relief=RIDGE,bd=2).grid(row=3,column=3,padx=2,pady=2)
btn_div = Button(root, text = "/",padx=30,pady=15 , 
                 command = div,relief=RIDGE,bd=2).grid(row=4,column=3,padx=2,pady=2)
btn_clear = Button(root, text = "C",padx=30,pady=15 ,
                   command = clear,relief=RIDGE,bd=2).grid(row=4,padx=2,pady=2,sticky=W+E)
btn_equal = Button(root, text = "=",padx=30,pady=15 ,
                   command = equal,relief=RIDGE,bd=2).grid(row=4,column=2,padx=2,pady=2)


root.mainloop()
```
